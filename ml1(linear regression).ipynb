```python
from sklearn.datasets import fetch_openml

# Load the Boston housing dataset
boston_df = fetch_openml(name='boston')

# Access the data and target
X, y = boston_df.data, boston_df.target
```

```python
df=pd.DataFrame(boston_df.data)
df
```

```python
df['price']=boston_df.target    #added dependent feature
```

```python
df
```

```python
#Dividing the datasets into independent and dependent features
x=df.iloc[:,:-1]            #skippping last feature - independent feature
y=df.iloc[:,-1]             #taking only last column - dependent feature
```

```python
x['CHAS'] = df['CHAS'].astype(float)
x['RAD'] = df['RAD'].astype(float)
```

```python
y.dtypes
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
lin_reg=LinearRegression()
mse=cross_val_score(lin_reg,x,y,scoring="neg_mean_squared_error",cv=5)
print(mse)                 #5 output because 5 cross validaion is given
```

```python
mean_of_mse=np.mean(mse)
print(mean_of_mse)
```

```python
lin_reg.fit(x,y)
print(lin_reg.predict(x))              #lin_req.predict() - give the value inside predict to make prediction
```

```python
#ridge & lasso regression
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()

params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,50,70,100]}

ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=10)
ridge_regressor.fit(x,y)
```

```python
x
```

```python
y
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)    #ridge regression is better as it is closest of all to 0
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()

params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,50,70,100]}

lasso_regressor=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=10)
lasso_regressor.fit(x,y)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```
